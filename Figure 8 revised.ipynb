```python
import pandas as pd
import matplotlib.pyplot as plt
import polyan
```

```python
#comparison for GSE13750: rich medium vs amino acid starvation
rich = pd.read_csv('Data/GSE13750.csv')
noaa = pd.read_csv('Data/GSE13750_noaa.csv')
r_poly = polyan.fp2poly(rich,include_idle = False)
r_poly_trace = polyan.plot_poly(r_poly)
noaa_poly = polyan.fp2poly(noaa,include_idle = False)
noaa_poly_trace = polyan.plot_poly(noaa_poly)
fig,ax = plt.subplots()
ax.scatter(r_poly_trace[0],r_poly_trace[1],c='black',s=2)
ax.scatter(noaa_poly_trace[0],noaa_poly_trace[1],c='red',s=2)
ax.set_yticks(())
ax.set_xticks(())
ax.set_ylabel('oD254')
ax.set_xlabel('Gradient')
ax.legend(['aa replete','aa starved'])
#plt.savefig('Figure 8.svg')
plt.show()
```
